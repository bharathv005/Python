## Hierarchical Clustering

- Hierarchical clustering is a method of cluster analysis which seeks to build a hierarchy of clusters.
- Hierarchical clustering is a popular method for visualizing the structure of a dataset.
- It is also a useful tool for making predictions about the membership of new observations.
- It is a bottom-up approach: each observation starts in its own cluster, and pairs of clusters are merged as one moves up the hierarchy.
- The results of hierarchical clustering are usually presented in a dendrogram.
- The dendrogram is a tree diagram showing the arrangement of the clusters produced by the algorithm.
- The height on the dendrogram at which two clusters are joined represents the similarity of the clusters.

##  Important Terms in Hierarchical Clustering
### Agglomerative Hierarchical Clustering
- Agglomerative hierarchical clustering is a type of hierarchical clustering algorithm.
- It starts with each data point as a separate cluster and then combines the two most similar clusters until all clusters have been combined into one.

### Divisive Hierarchical Clustering
- Divisive hierarchical clustering is a type of hierarchical clustering algorithm.
- It starts with all data points in one cluster and then splits the cluster into two until each cluster contains only one data point.

### Dendrogram
- A dendrogram is a tree diagram that records the sequences of merges or splits in a hierarchical clustering.
- The height on the dendrogram at which two clusters are joined represents the similarity of the clusters.

### Linkage Criteria
- Linkage criteria are used to determine the distance between clusters.
- There are four linkage criteria: single, complete, average, and centroid.
- The single linkage criteria uses the distance between the closest points in the clusters.
- The complete linkage criteria uses the distance between the farthest points in the clusters.
- The average linkage criteria uses the average distance between all points in the clusters.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
stocks = ['AAPL', 'AMZN', 'AXP', 'COP', 'COST', 'CSCO', 'CVX',
         'GOOGL', 'HD', 'INTC', 'JPM', 'MSFT', 'SLB', 'TGT',
         'USB', 'WFC', 'WMT', 'XOM']

df = pd.read_csv('data/sp500_data.csv', index_col=0)

df = df[stocks].T
df.head()
```

```python
Z = linkage(df, 'complete')
plt.figure(figsize=(12, 5), dpi=200)
dendrogram(Z, labels=df.index, leaf_rotation=90, leaf_font_size=16)
plt.title('Hierarchical clustering')
plt.xlabel('sample index')
plt.ylabel('distance')
plt.show()
```
