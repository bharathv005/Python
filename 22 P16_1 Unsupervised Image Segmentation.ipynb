## Unsupervised Image Segmentation

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image

from sklearn.cluster import KMeans


# Read in the image
image = Image.open('data/flower.jpg')
# image

image = np.array(image)
image.shape

X = image.reshape(-1, 3)

kmenas = KMeans(n_clusters=3, random_state=42)
kmenas.fit(X)

# get the segmented image
segmented_img = kmenas.cluster_centers_[kmenas.labels_]
segmented_img = segmented_img.reshape(image.shape)

plt.imshow(segmented_img/255)
```

```python
segmented_img = []
max_color = 7
for n_clusters in range(1, max_color):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    segmented_img.append(kmeans.cluster_centers_[kmeans.labels_].reshape(image.shape))


# Plot the results
fig, axes = plt.subplots(2, 3, figsize=(15, 8), sharex=True, sharey=True)
ax = axes.ravel()

for n_clusters in range(1, max_color):
    ax[n_clusters-1].imshow(segmented_img[n_clusters-1]/255)
    ax[n_clusters-1].set_title('n_clusters = {}'.format(n_clusters))

ax[0].set_xticks(())
ax[0].set_yticks(())
plt.tight_layout()
plt.show()
```
