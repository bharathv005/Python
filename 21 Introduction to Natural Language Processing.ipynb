### What are Key NLP Techniques


- Tokenization
    - Tokenization is the process of breaking a stream of text up into words, phrases, symbols, or other meaningful elements called tokens. For example, the sentence The cow jumped over the moon. would be tokenized as [The, cow, jumped, over, the, moon, .]. The process of tokenization is useful for a number of reasons. For example, it allows us to easily determine the vocabulary size of a corpus, and it allows us to easily compute statistics on the occurrence of individual words in a corpus.
- Stemming
    - Stemming is the process of reducing inflected (or sometimes derived) words to their stem, base or root form. For example, the words play, player, played, plays, playing have the common stem play.
- Lemmatization
    - Lemmatization is the process of grouping together the inflected forms of a word so they can be analysed as a single item. For example, the words play, player, played, plays, playing have the common stem play.
- Stop Words
    - Stop words are words which are filtered out before or after processing of natural language data (text). Though stop words usually refers to the most common words in a language, there is no single universal list of stop words used by all natural language processing tools, and indeed not all tools even use such a list. Some tools specifically avoid removing these stop words to support phrase search.
- Bag of Words
    - Bag of words is a representation of text that describes the occurrence of words within a document. It involves two things:
        - A vocabulary of known words.
        - A measure of the presence of known words.
- TF-IDF
    - TF-IDF is a statistical measure that evaluates how relevant a word is to a document in a collection of documents. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus.
- Word Embedding
    - Word embedding is a technique of natural language processing where words or phrases from the vocabulary are mapped to vectors of real numbers. Conceptually it involves a mathematical embedding from a space with one dimension per word to a continuous vector space with a much lower dimension.
    - Word2Vec
    - GloVe
    - FastText

### Common Challanges in NLP
- **Sparsity of data**
    - The data is sparse, meaning that there are not enough examples to learn from. This is especially true for languages with small vocabularies, such as English. For example, the English language has 26 letters, but only 44 sounds. This means that there are many words that sound the same but are spelled differently. For example, cat and hat are pronounced the same, but they are spelled differently. This is a problem because the model will not be able to learn the relationship between the two words.
- **Ambiguity**
    - Ambiguity is a problem in NLP because it is difficult to determine the meaning of a word in a sentence. For example, bank account and river bank
- **Irrelevant features**
    - Irrelevant features are features that are not relevant to the task at hand. For example, stop words.
- **Out of vocabulary (OOV)**
    - Out of vocabulary (OOV) is a problem in NLP because it is difficult to determine the meaning of a word in a sentence.
- **Noise**
    - Noise is a problem in NLP because it is difficult to determine the meaning of a word in a sentence.

### Bag of Words - The Simples Word Embedding Technique

Bag of Words (BoW) is a technique used in natural language processing (NLP) to represent a text document as a numerical vector. The basic idea behind BoW is to convert a text document into a vector of word counts, where each dimension of the vector corresponds to a specific word in the vocabulary.

Process to create a BoW model:
- **Step 1:** Create a vocabulary of unique words from the entire corpus.
- **Step 2:** Create a vector of size equal to the vocabulary size. Initialize all the values to 0.
- **Step 3:** For each word in the document, increment the count of the corresponding index in the vector by 1.
- **Step 4:** Repeat steps 2 and 3 for all the documents in the corpus.

### Term Frequency - Inverse Document Frequency (TF-IDF)
https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfTransformer.html

TF-IDF (Term Frequency-Inverse Document Frequency) is a statistical measure that is used to evaluate the importance of a word in a document or a corpus of documents. It is calculated by multiplying two metrics: term frequency (TF) and inverse document frequency (IDF).
- TF is the frequency of a word in a document. It is calculated as the number of times a word appears in a document divided by the total number of words in the document.
- IDF is the inverse of the document frequency. It is calculated as the logarithm of the number of documents divided by the number of documents that contain the word.
- TF-IDF is the product of TF and IDF. It is used to determine the importance of a word in a document.

**Example of TF-IDF**
- Consider a document containing 100 words wherein the word cat appears 3 times.
- The term frequency (i.e., tf) for cat is then (3 / 100) = 0.03.
- Now, assume we have 1000 documents and the word cat appears in 10 documents. Then, the inverse document frequency (i.e., idf) is calculated as log(1000/10) = 2.
- Thus, the Tf-idf weight is the product of these quantities: 0.03 * 2 = 0.06.

#### Train Test Split

````python
from sklearn.model_selection import train_test_split
````

````python
X_train,X_test,y_train,y_test=train_test_split(df['message'], df['label'],
                                              test_size=0.2, random_state=0,
                                              stratify=df['label'])
````

````python
X_train.shape, X_test.shape
````

### TF-IDF

````python
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
tfidf = TfidfVectorizer()
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
````

````python
len(tfidf.get_feature_names())
````

````python
X_train.toarray().shape
````

### Model Building

````python
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
````

````python
svm = LinearSVC()
svm.fit(X_train, y_train)
````

````python
y_pred = svm.predict(X_test)
````

````python
print(classification_report(y_test, y_pred))
````

````python
text = ['hey whassup?']
text = tfidf.transform(text)
text.toarray().shape

svm.predict(text)
````

````python
text = ['you have won tickets to the USA this summer']
text = tfidf.transform(text)
text.toarray().shape

svm.predict(text)
````
