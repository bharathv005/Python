### DBSCAN

DBSCAN is a clustering algorithm that is based on density of points in a region. It groups together points that are closely packed together (points with many nearby neighbors), marking as outliers points that lie alone in low-density regions (whose nearest neighbors are too far away).

DBSCAN is one of the most common clustering algorithms and also works well in complex domains where other clustering algorithms fail.

[DBSCAN](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html)

### DBSCAN Algorithm
- Choose an arbitrary point that has not been visited. This point is called the current point.
- Find all the points within a distance $Eps$ of the current point. If there are at least MinPts such points, a cluster is started. Add to this cluster all points found within a distance $Eps$ of any point in this cluster. This can be done iteratively, i.e. the new points found within $Eps$ of any point in the cluster are added to the cluster as well. Continue until no new points can be added.
- Choose a new unvisited point and repeat the process.
- Any point that was not added to any cluster is marked as noise.



### DBSCAN Parameters
- $Eps$ - The maximum distance between two samples for them to be considered as in the same neighborhood.
- MinPts - The number of samples (or total weight) in a neighborhood for a point to be considered as a core point. This includes the point itself.

### DBSCAN Pros
- DBSCAN is robust to outliers.
- DBSCAN is able to find clusters of arbitrary shape.
- DBSCAN is able to find clusters of different sizes and densities.

### DBSCAN Cons
- DBSCAN is not suitable for data which contains clusters of similar density.
- DBSCAN requires two parameters, $Eps$ and MinPts, which are not easy to define.
- DBSCAN is not suitable for high dimensional data.



### Data Generation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
```

```python
X,y = make_moons(n_samples=200, noise=0.1, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

### DBSCAN Coding

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
labels = kmeans.labels_

# scatter plot
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title('K-Means Clustering')
```

```python
# DBSCAN Coding
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
clusters = dbscan.fit_predict(X)
# clusters
# scatter plot
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.title('DBSCAN Clustering')
```

## Spectral Clustering
Specral clustering is a clustering algorithm based on the eigenvectors of the Laplacian matrix of the graph of pairwise similarities between the samples. It is particularly effective if the clusters are globular.

[Specral Clustering](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.SpectralClustering.html)



### Spectral Clustering Algorithm
- Construct a similarity graph from the data.
- Compute the normalized Laplacian of the graph.
- Compute the eigenvectors of the normalized Laplacian.
- Use k-means to cluster the eigenvectors.

### Spectral Clustering Pros
- Spectral clustering is very efficient.
- Spectral clustering is very flexible in terms of the affinity matrix that can be used to construct the graph of similarities between samples.

### Spectral Clustering Cons
- Spectral clustering is not scalable to large number of samples.
- Spectral clustering requires the number of clusters to be specified.
- Spectral clustering does not work well when clusters have non-convex shapes.


```python
from sklearn.cluster import SpectralClustering
```

```python
sc = SpectralClustering(n_clusters=2, gamma=100)
sc.fit(X)

y_sc = sc.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:,1], c=y_sc)
plt.title('Spectral Clustering')
plt.show()
```
